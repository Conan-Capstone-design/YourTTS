```python
import os
import csv

base_path = "/home/s_20221326/TTS_dataset"  # 전체 데이터셋 경로
speaker = "conan"
speaker_path = os.path.join(base_path, speaker)
output_csv = os.path.join(base_path, "metadata_conan.csv")

with open(output_csv, "w", encoding="utf-8", newline="") as f:
    writer = csv.writer(f, delimiter="|")
    for file in os.listdir(speaker_path):
        if file.endswith(".wav"):
            base_name = os.path.splitext(file)[0]
            wav_path = f"{speaker}/{base_name}.wav"
            txt_path = os.path.join(speaker_path, f"{base_name}.txt")

            if os.path.exists(txt_path):
                with open(txt_path, "r", encoding="utf-8") as tf:
                    transcript = tf.read().strip()
                writer.writerow([wav_path, transcript, speaker])

print(f"metadata_conan.csv 생성 완료: {output_csv}")
```

```python
import os

import torch
from trainer import Trainer, TrainerArgs

from TTS.bin.compute_embeddings import compute_embeddings
from TTS.bin.resample import resample_files
from TTS.config.shared_configs import BaseDatasetConfig
from TTS.tts.configs.vits_config import VitsConfig
from TTS.tts.datasets import load_tts_samples
from TTS.tts.models.vits import CharactersConfig, Vits, VitsArgs, VitsAudioConfig
from TTS.utils.downloaders import download_vctk

torch.set_num_threads(24)

# pylint: disable=W0105
"""
    This recipe replicates the first experiment proposed in the YourTTS paper (https://arxiv.org/abs/2112.02418).
    YourTTS model is based on the VITS model however it uses external speaker embeddings extracted from a pre-trained speaker encoder and has small architecture changes.
    In addition, YourTTS can be trained in multilingual data, however, this recipe replicates the single language training using the VCTK dataset.
    If you are interested in multilingual training, we have commented on parameters on the VitsArgs class instance that should be enabled for multilingual training.
    In addition, you will need to add the extra datasets following the VCTK as an example.
"""
CURRENT_PATH = os.getcwd()

# Name of the run for the Trainer
RUN_NAME = "YourTTS-SingleSpeaker-conan"

OUT_PATH = os.path.join(CURRENT_PATH, "output6")  # "/raid/coqui/Checkpoints/original-YourTTS/"

# If you want to do transfer learning and speedup your training you can set here the path to the original YourTTS model
RESTORE_PATH = "/home/s_20221326/output2/yourtts-multichar-June-05-2025_12+06AM-0000000/best_model.pth" #"/home/s_20221326/TTS/pretrained/model/best_model_YourTTS_VCTK.pth"  # "/root/.local/share/tts/tts_models--multilingual--multi-dataset--your_tts/model_file.pth"

# This paramter is useful to debug, it skips the training epochs and just do the evaluation  and produce the test sentences
SKIP_TRAIN_EPOCH = False

# Set here the batch size to be used in training and evaluation
BATCH_SIZE = 52

# Training Sampling rate and the target sampling rate for resampling the downloaded dataset (Note: If you change this you might need to redownload the dataset !!)
# Note: If you add new datasets, please make sure that the dataset sampling rate and this parameter are matching, otherwise resample your audios
SAMPLE_RATE = 16000

# Max audio length in seconds to be used in training (every audio bigger than it will be ignored)
MAX_AUDIO_LEN_IN_SECONDS = 20
DATASET_PATH = "/home/s_20221326/TTS_dataset"
METADATA_FILE = "metadata_conan.csv"  # wav|text|speaker 형식이어야 함
SPEAKER_NAME = "conan"  # 단일 화자 이름

from TTS.tts.datasets.formatters import bel_tts_formatter
#  데이터셋 구성
dataset_config = BaseDatasetConfig(
    formatter="bel_tts_formatter",  # 또는 너의 포맷터
    dataset_name=SPEAKER_NAME,
    path=DATASET_PATH,
    meta_file_train=METADATA_FILE,
    language="ko",
)
```

```python
# 완성형 한글 (가 ~ 힣)
hangul_complete = [chr(c) for c in range(ord('가'), ord('힣') + 1)]

# 추가 기호
symbols = list("ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz0123456789!?.,~ '")

# 문자 리스트 합치기
all_chars = hangul_complete + symbols

# 저장할 경로
save_path = "/home/s_20221326/YourTTS/characters.json"

# 저장
char_dict = {"characters": ''.join(all_chars)}
with open(save_path, "w", encoding="utf-8") as f:
    json.dump(char_dict, f, ensure_ascii=False, indent=2)

print(f"저장 위치 : {save_path}")
```

```python
# Audio config used in training.
audio_config = VitsAudioConfig(
    sample_rate=SAMPLE_RATE,
    hop_length=256,
    win_length=1024,
    fft_size=1024,
    mel_fmin=0.0,
    mel_fmax=None,
    num_mels=80,
)

# Init VITSArgs setting the arguments that are needed for the YourTTS model
model_args = VitsArgs(
    use_d_vector_file=True,
    d_vector_file=["/home/s_20221326/YourTTS/embedding/speakers_conan.pth"],  # 리스트 형식!
    d_vector_dim=512,
    speaker_encoder_model_path="/home/s_20221326/YourTTS/pretrained/se_model/model_se.pth",
    speaker_encoder_config_path="/home/s_20221326/YourTTS/pretrained/se_model/config_se.json",
    num_speakers=1,
    num_languages=1
    # In the paper, we accidentally trained the YourTTS using ResNet blocks type 2, if you like you can use the ResNet blocks type 1 like the VITS model
    # Useful parameters to enable the Speaker Consistency Loss (SCL) described in the paper
    # use_speaker_encoder_as_loss=True,
    # Useful parameters to enable multilingual training
    # use_language_embedding=True,
    # embedded_language_dim=4,
)
with open("/home/s_20221326/YourTTS/characters.json", "r", encoding="utf-8") as f:
    char_dict = json.load(f)

# General training config, here you can change the batch size and others useful parameters
config = VitsConfig(
    output_path=OUT_PATH,
    model_args=model_args,
    epochs=500,
    run_name=RUN_NAME,
    project_name="YourTTS_conan",
    run_description="""
            - Original YourTTS trained using VCTK dataset
        """,
    dashboard_logger="tensorboard",
    logger_uri=None,
    audio=audio_config,
    batch_size=BATCH_SIZE,
    lr = 0.01,
    grad_clip= [1000, 1000],
    batch_group_size=48,
    eval_batch_size=BATCH_SIZE,
    num_loader_workers=8,
    eval_split_max_size=256,
    print_step=50,
    plot_step=100,
    log_model_step=1000,
    kl_loss_alpha=1.0,
    mel_loss_alpha = 45.0,
    save_step=5000,
    save_n_checkpoints=2,
    disc_loss_alpha = 1.0,
    feat_loss_alpha = 1.0,
    save_checkpoints=True,
    target_loss="loss_1",
    print_eval=False,
    use_phonemes=False,
    phonemizer="espeak",
    phoneme_language="ko",
    compute_input_seq_cache=True,
    add_blank=True,
    text_cleaner=False,
    characters=CharactersConfig(
        characters_class="TTS.tts.models.vits.VitsCharacters",
        characters=char_dict["characters"],  # 문자열로 넘겨줌
        punctuations="!?.,~ ",
        pad="_",
        eos="&",
        bos="*",
        blank=None,
        is_unique=True,
        is_sorted=True
    ),
    phoneme_cache_path=None,
    precompute_num_workers=12,
    start_by_longest=True,
    datasets=[dataset_config],
    cudnn_benchmark=False,
    max_audio_len=SAMPLE_RATE * MAX_AUDIO_LEN_IN_SECONDS,
    mixed_precision=False,
    test_sentences=[
        ["꽤 오랜 시간이 걸렸지만, 이제 내 목소리를 찾았으니 침묵하지 않을 거예요."],
        ["메아리가 아니라, 목소리가 되어야 해요."],
        ["죄송합니다, 데이브. 그건 할 수 없어요."],
        ["이 케이크 정말 맛있고 촉촉하네요."],
        ["1963년 11월 22일 이전의 이야기입니다."]
    ],
    # Enable the weighted sampler
    use_weighted_sampler=False,

    # It defines the Speaker Consistency Loss (SCL) α to 9 like the paper
    speaker_encoder_loss_alpha=9.0,
)

# Load all the datasets samples and split traning and evaluation sets
train_samples, eval_samples = load_tts_samples(
    config.datasets,
    eval_split=True,
    eval_split_max_size=128,
    formatter=bel_tts_formatter  # ← 이걸 추가
)


# Init the model
model = Vits.init_from_config(config)

# Init the trainer and 🚀
trainer = Trainer(
    TrainerArgs(restore_path=RESTORE_PATH, skip_train_epoch=SKIP_TRAIN_EPOCH),
    config,
    output_path=OUT_PATH,
    model=model,
    train_samples=train_samples,
    eval_samples=eval_samples,
)

```

```python
trainer.fit()
```
